**CTD Processing**

Para utilizar as funções do projeto CTD Processing, primeiro coloque o arquivo 'CTD_utilities.py' no diretório desejado.

Em um novo script python, importe todo o conteúdo do módulo CTD_utilities:

```python
from CTD_utilities import *
```

Depois disso já é possível chamar as funções para fazer o processamento do seus dados.

**Funções disponíveis até a versão mais atual (1.7.0)**




*   **plot**: está função plota um simples perfil de temperatura
*   data: arquivo de dados
*   path: nome e o caminho que a imagem gerada será salva



```python
plot(data, path)
```



*   **convert**: Para cada coluna do arquivo, exceto a coluna 'time', substitui ',' por '.', e nas células que tenham o '.' como separador de milhar, retira.
*   data: arquivo de dados



```python
convert(data)
```



*   **downcast**: retira os dados coletados durante o upcast
*   data: arquivo de dados



```python
downcast(data)
```



*   **remove_outliers**: retira os valores considerados outliers utilizando o método 3-sigma
*   data: arquivo de dados



```python
remove_outliers(data)
```



*   **above_sea_level**: retira os dados que possam ter sido coletados acima do nível do mar
*   data: arquivo de dados
*   sea_level_pressure: valor em dBar do nível do mar (normalmente é 10.12)



```python
above_sea_level(data, sea_level_pressure)
```



*   **pressure_loops**: remove os loops de pressão dos dados
*   data: arquivo de dados



```python
pressure_loops(data)
```



*   **lp_filter**: filtro passa-baixa
*   data: arquivo de dados
*   sample_rate: taxa de amostragem
*   time_constant: constante de tempo



```python
lp_filter(data, sample_rate, time_constant)
```



*   **bin_average**: organiza os dados em bins pela média
*   data: arquivo de dados
*   tamanho_janela: de quantos em quantos metros de profundidade calcular os bins
*   coluna: coluna de dados para a binagem
*   coluna2: segunda coluna de dados para a binagem
*   lat: latitude



```python
bin_average(data, tamanho_janela, coluna, coluna2, lat)
```



*   **plot_perfil_termosal**: plotter de um perfil de temperatura e salinidade sobrepostos
*   data: arquivo de dados
*   path: nome e o caminho que a imagem gerada será salva



```python
plot_perfil_termosal(data, path)
```



*   **diagramats**: plotter de um diagram T-S com isolinhas de densidade
*   data: arquivo de dados
*   path: nome e o caminho que a imagem gerada será salva
*   lon: longitude
*   lat: latitude



```python
diagramats(data, path, lon, lat)
```
